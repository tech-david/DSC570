```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import CategoricalNB, GaussianNB
from sklearn import metrics
import warnings
warnings.filterwarnings('ignore')
```

```python
performance_decision=pd.read_csv('PerformanceDecision.csv',
                                header=[0])
performance_decision
```

```python
## Setting salary as target variable
y = performance_decision[['status']]
## Setting all other variables except salary as independent
X = performance_decision.loc[:,
                            performance_decision.columns != 'status']
y
```

```python
X
```

```python
## encoding status, age, and salary to ordinal attributes
encoder = OrdinalEncoder()
status_encoded = encoder.fit_transform(y[['status']])
y[['status']] = status_encoded
salary_encoded = encoder.fit_transform(X[['salary']])
X[['salary']] = salary_encoded
age_encoded = encoder.fit_transform(X[['age']])
X[['age']] = age_encoded


## encoding department to numerical attribute
label_encoder = LabelEncoder()
department_encoded = label_encoder.fit_transform(X[['department']])
X[['department']] = department_encoded
X
```

```python
## Train-test spit
X_train, X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   test_size=0.25)
```

```python
## Categorical Naive Bayes implementation
category_naive_bayes = CategoricalNB()
category_naive_bayes.fit(X_train,y_train)
y_pred = category_naive_bayes.predict(X_test)
print("Accuracy: ",
     metrics.accuracy_score(y_test,
                           y_pred))
```

```python
## Gaussain Naive Bayes implementation
guassianNB = GaussianNB()
guassianNB.fit(X_train, y_train)
y_gnb_pred =guassianNB.predict(X_test)
print("Accuracy: ",
     metrics.accuracy_score(y_test,
                           y_gnb_pred))
```
