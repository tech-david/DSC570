```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
performance_decision=pd.read_csv('PerformanceDecision.csv',
                                header=[0])
performance_decision
```

```python
sns.set_theme(style='darkgrid')
```

```python
sns.barplot(x=performance_decision['department'],
           y=performance_decision['salary'],
           hue=performance_decision['status']);
plt.savefig("department-salary-by-status.png")
```

```python
sns.barplot(x=performance_decision['status'],
           y=performance_decision['salary'],
           hue=performance_decision['age']);
plt.savefig("status-salary-by-age.png")
```

```python
sns.barplot(x=performance_decision['status'],
           y=performance_decision['salary'],
           hue=performance_decision['department']);
plt.savefig("status-salary-by-department.png")
```

```python
sns.barplot(x=performance_decision['department'],
           y=performance_decision['salary']);
plt.savefig("department-salary.png")
```

```python
sns.barplot(x=performance_decision['status'],
           y=performance_decision['age']);
plt.savefig("status-age.png")
```

```python
sns.barplot(x=performance_decision['status'],
           y=performance_decision['salary']);
plt.savefig("status-salary.png")
```

```python
sns.barplot(x=performance_decision['department'],
           y=performance_decision['salary'],
           hue=performance_decision['age']);
plt.savefig("department-salary-by-age.png")
```

```python
sns.catplot(data=performance_decision,
            x='department',
            y='salary',
            col='status',
            kind='bar');
plt.savefig("compare-status-plots-department-salary.png")
```

```python
sns.pairplot(data=performance_decision,
            hue='department');
plt.savefig("pairplot-data.png")
```
